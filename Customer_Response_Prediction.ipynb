```python
df['Income']=df['Income'].fillna(df['Income'].median())
```

```python
# converting datetime assuming day-first format
df["Dt_Customer"] = pd.to_datetime(df["Dt_Customer"],dayfirst=True)

# Getting today's date
today = pd.Timestamp.today().normalize()

# calculating tenure in days
df["Customer_Tenure_Days"] = (today - df["Dt_Customer"]).dt.days

#Dropping Dt_Customer column as we have already calculated Tenure Days.
df=df.drop(columns=['Dt_Customer'])
```

```python
target='Response'
df_numeric=df.drop(['Education','Marital_Status'],axis=1)
corr=df_numeric.corr()[target]
print(corr)
```

```python
#Checking Education's effect on response
education_type_response=df.groupby("Education")["Response"].sum()
education_type_response
```

```python
#Checking Marital status's effect on response
Marital_Status_response=df.groupby("Marital_Status")["Response"].sum()
Marital_Status_response
```

```python
#One hot encoding for Categorical columns Education and Marital Status(One hot encoding)
df=pd.get_dummies(df,columns=['Education','Marital_Status'],drop_first=True)
```

```python
#Aways check whether it is imbalanced data, If it is imbalanced then even simple logistic regression gives more accuracy, but it is meaningless
df['Response'].value_counts()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

X=df.drop(['Response','Z_Revenue','Z_CostContact'],axis=1)#Removing Z_Revenue and Z_CostContact as they dont have any correlation effect with Response
Y=df['Response']

X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=42)

"""We use fit_transform() on the training data because the scaler must learn the mean and standard deviation only from the training set. 
These learned parameters are then applied to both the training and test data.If we fit the scaler on the test data, the mean and standard
deviation would be computed using test information, which leads to data leakage and results in an optimistically biased and invalid model evaluation.
"""
scaler=StandardScaler()
X_train_scaled=scaler.fit_transform(X_train)#train date is using fit_tran
X_test_scaled=scaler.transform(X_test) #

```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV

# model
model = LogisticRegression()

# parameter grid for class weights, Different C values are used for Regularization.
param_grid = { "C":[0.01,0.1,1,10],"class_weight": [ None, "balanced",{0: 1, 1: 2},{0: 1, 1: 3},{0: 1, 1: 5}]}

# grid search
grid = GridSearchCV(estimator=model,param_grid=param_grid,scoring="f1")  # f1 score is good metric imbalanced data

grid.fit(X_train_scaled,Y_train)

best_model=grid.best_estimator_

```

```python
#Predicting
Y_pred=best_model.predict(X_test_scaled)

from sklearn.metrics import classification_report,roc_auc_score,precision_score,recall_score

#Accuracy is not used because for imbalanced data accuracy is not a good metric for evaluating model
precision=precision_score(Y_test,Y_pred)
print("precision=",precision)

recall=recall_score(Y_test,Y_pred)
print("Recall=",recall)

Y_prob=best_model.predict_proba(X_test_scaled)[:,1]
ROC_AUC_score=roc_auc_score(Y_test,Y_prob)
print("ROC_AUC_score",ROC_AUC_score)
```

```python
#As ROC_AUC score is good, model is discriminating positive and negative classes well, Now issue is threshold.
#Increasing threshold to make precision score better , Here we are mainly focusing on avoiding false positives 
#This scenario works better when company dont want to  target on cusomers who are unlikely to respond.

Y_prob=best_model.predict_proba(X_test_scaled)[:,1]

#Here we are mainly focucing on not missing out responders
threshold = 0.70
Y_pred = (Y_prob>=threshold).astype(int)

from sklearn.metrics import classification_report,roc_auc_score,precision_score,recall_score

#Accuracy is not used because for imbalanced data accuracy is not a good metric for evaluating model
precision=precision_score(Y_test,Y_pred)
print("precision=",precision)

recall=recall_score(Y_test,Y_pred)
print("Recall=",recall)

Y_prob=best_model.predict_proba(X_test_scaled)[:,1]
ROC_AUC_score=roc_auc_score(Y_test,Y_prob)
print("ROC_AUC_score",ROC_AUC_score)
```

```python
#As ROC_AUC score is good, model is discriminating positive and negative classes well, Now issue is threshold.
#Decreasing threshold to make recall score better 
#This scenario works better when company focusing on not missing potential responders.

Y_prob=best_model.predict_proba(X_test_scaled)[:,1]

#Here we are mainly focucing on not missing out responders
threshold = 0.30
Y_pred = (Y_prob>=threshold).astype(int)

from sklearn.metrics import classification_report,roc_auc_score,precision_score,recall_score

#Accuracy is not used because for imbalanced data accuracy is not a good metric for evaluating model
precision=precision_score(Y_test,Y_pred)
print("precision=",precision)

recall=recall_score(Y_test,Y_pred)
print("Recall=",recall)

Y_prob=best_model.predict_proba(X_test_scaled)[:,1]
ROC_AUC_score=roc_auc_score(Y_test,Y_prob)
print("ROC_AUC_score",ROC_AUC_score)

```
