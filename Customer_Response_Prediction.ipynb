```python
df['Income']=df['Income'].fillna(df['Income'].median())
```

```python
#Aways check whether it is imbalanced data, If it is imbalanced then even simple logistic regression gives more accuracy, but it is meaningless

df['Response'].value_counts()
```

```python
df.columns
```

```python
#One hot encoding for Education and Marital Status(One hot encoding is prefered for PCA as distance matters here
df=pd.get_dummies(df,columns=['Education','Marital_Status'],drop_first=True)
```

```python
#Convert Dt_Customer column to proper datetime and extract year, month
df['Dt_Customer']=pd.to_datetime(df['Dt_Customer'],format="mixed")
df['Customer_Year']=df['Dt_Customer'].dt.year
df['Customer_Month']=df['Dt_Customer'].dt.month


```

```python
#Dropping Dt_Customer column as we have already extracted Customer_Year and Customer_Month columns from it
df=df.drop(columns=['Dt_Customer'])
```

```python
from sklearn.model_selection import train_test_split

X=df.drop('Response',axis=1)#Make sure all features are in numeric format before assigning
Y=df['Response']

X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=42)

from sklearn.preprocessing import StandardScaler

scaler=StandardScaler()
X_train_scaled=scaler.fit_transform(X_train)
X_test_scaled=scaler.transform(X_test)

```

```python
from sklearn.linear_model import LogisticRegression
from imblearn.over_sampling import SMOTE

smote=SMOTE(random_state=42)
X_res,Y_res=smote.fit_resample(X_train_scaled,Y_train)
print("Before",Y_train.value_counts()) 
print("After",Y_res.value_counts())

# Train Logistic Regression model
model = LogisticRegression()
model.fit(X_res, Y_res)

Y_pred=model.predict(X_test_scaled)

from sklearn.metrics import accuracy_score,classification_report,roc_auc_score,precision_score,recall_score

Train_Accuracy=model.score(X_train_scaled,Y_train)
print("Train_Accuracy",Train_Accuracy)

Test_accuracy=accuracy_score(Y_test,Y_pred)
print("Test accuracy=",Test_accuracy)

precision=precision_score(Y_test,Y_pred)
print("precision=",precision)

recall=recall_score(Y_test,Y_pred)
print("Recall=",recall)

ROC_AUC_score=roc_auc_score(Y_test,model.predict_proba(X_test_scaled)[:,1])
print("ROC_AUC_score",ROC_AUC_score)
```

```python
from sklearn.ensemble import RandomForestClassifier
from imblearn.over_sampling import SMOTE

smote=SMOTE(random_state=42)
X_res,Y_res=smote.fit_resample(X_train_scaled,Y_train)
print("Before",Y_train.value_counts()) 
print("After",Y_res.value_counts())

model=RandomForestClassifier(n_estimators=200,max_depth=4,min_samples_split=5,min_samples_leaf=5,max_features='sqrt',random_state=42)
model.fit(X_res,Y_res)

Y_pred=model.predict(X_test_scaled)

from sklearn.metrics import accuracy_score,classification_report,roc_auc_score,precision_score,recall_score

Train_Accuracy=model.score(X_train_scaled,Y_train)
print("Train_Accuracy",Train_Accuracy)

Test_accuracy=accuracy_score(Y_test,Y_pred)
print("Test accuracy=",Test_accuracy)

precision=precision_score(Y_test,Y_pred)
print("precision=",precision)

recall=recall_score(Y_test,Y_pred)
print("Recall=",recall)

ROC_AUC_score=roc_auc_score(Y_test,model.predict_proba(X_test_scaled)[:,1])
print("ROC_AUC_score",ROC_AUC_score)
```

```python
from sklearn.model_selection import train_test_split
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score,classification_report
from imblearn.over_sampling import SMOTE

smote=SMOTE(random_state=42)
X_res,Y_res=smote.fit_resample(X_train_scaled,Y_train)
print("Before",Y_train.value_counts()) 
print("After",Y_res.value_counts())


model=XGBClassifier(n_estimators=80,learning_rate=0.01,max_depth=3,subsample=0.4,colsample_bytree=0.4,reg_lambda=0.1,reg_alpha=0,gamma=0.0,random_state=42)
model.fit(X_res,Y_res)

Y_pred=model.predict(X_test_scaled)

from sklearn.metrics import accuracy_score,classification_report,roc_auc_score,precision_score,recall_score

Train_Accuracy=model.score(X_train_scaled,Y_train)
print("Train_Accuracy",Train_Accuracy)

Test_accuracy=accuracy_score(Y_test,Y_pred)
print("Test accuracy=",Test_accuracy)

precision=precision_score(Y_test,Y_pred)
print("precision=",precision)

recall=recall_score(Y_test,Y_pred)
print("Recall=",recall)

ROC_AUC_score=roc_auc_score(Y_test,model.predict_proba(X_test_scaled)[:,1])
print("ROC_AUC_score",ROC_AUC_score)
```
